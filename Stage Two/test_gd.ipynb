```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from random import randint
```

```python
#Create dummy data

data = {
    'X' : [((i*3)-randint(a = -1, b = 1)) for i in range(10000)],
    'y' : [j for j in range(10000)]
}
df = pd.DataFrame(data)
```

```python
import GradientDecentSecondTry as gd
```

```python
X = df['X']
y = df['y']
```

```python
plt.figure()
plt.scatter(X,y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
model = gd.Regression()  # Create an instance without setting epochs
```
